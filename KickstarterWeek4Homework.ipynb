```python
import pandas as pd

Location = "kickstarter.xls"
df = pd.read_excel(Location)
df.head()
```

```python
df.columns = ['ID','name','category','main_category','currency','deadline','goal','launched','pledged','state','backers','country','usd pledged','usd_pledged_real','usd_goal_real']
df.head()
```

```python
## Drop 'currency', 'goal', 'pledged', and 'usd pledged' columns.
dfdropcol = df
dfdropcol = dfdropcol.drop('currency', axis=1)
dfdropcol = dfdropcol.drop('goal', axis=1)
dfdropcol = dfdropcol.drop('pledged', axis=1)
dfdropcol = dfdropcol.drop('usd pledged', axis=1)
dfdropcol.head()
```

```python
dfdropcol.describe()
```

```python
dfdropcol.dtypes
```

```python
pd.value_counts(dfdropcol['state'])
```

```python
## Drop rows where state = live and state = undefined (pledged > 0 but backers = 0).
dfdrop = dfdropcol
dfdrop = dfdrop.drop(dfdrop[dfdrop['state']=='live'].index)
dfdrop = dfdrop.drop(dfdrop[dfdrop['state']=='undefined'].index)
dfdrop = dfdrop.sort_values(by='usd_goal_real', ascending=0)
dfdrop.head()
```

```python
pd.value_counts(dfdrop['state'])
```

```python
import numpy as np
dfdrop['state_recat'] = np.where(dfdrop['usd_pledged_real']>=dfdrop['usd_goal_real'],'successful','not_successful')
dfdrop.head()
```

```python
dfdrop[dfdrop['state_recat']=='successful'].head()
## The largest successful goal was $1,000,000. Video games, technology, and product design appear
## to be the most successful types of projects.
```

```python
## Drop rows where state_recat = 'successful' and backers = 0 (state should have been classified as 'undefined').
dfdrop = dfdrop.drop(dfdrop[(dfdrop['state_recat']=='successful') & (dfdrop['backers']==0)].index)
dfdrop.head()
```

```python
print(len(dfdrop[(dfdrop['state']=='canceled') & (dfdrop['state_recat']=='successful')]))
```

```python
print(len(dfdrop[(dfdrop['state']=='suspended') & (dfdrop['state_recat']=='successful')]))
```

```python
## Create numeric variable to represent successful or not_successful.
def score_to_numeric(x):
    if x=='sucessful':
        return 1
    if x=='not_successful':
        return 0
```

```python
dfdrop['success_val'] = dfdrop['state_recat'].apply(score_to_numeric)
dfdrop.head()
```

```python
dfdrop_success_val = pd.get_dummies(dfdrop['state_recat'])
dfdrop_success_val.head()
```

```python
dfdrop_new = dfdrop.join(dfdrop_success_val)
dfdrop_new.head()
```

```python
## Create new column for pledged as percent of goal.

dfdrop_new['percent'] = 100*(dfdrop_new['usd_pledged_real'])/(dfdrop_new['usd_goal_real'])
dfdrop_new.head()
```

```python
dfdrop_new.describe()
```

```python
dfdrop_new.mode()
```

```python
dfdrop_new.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='successful ~ usd_goal_real + usd_pledged_real + backers', data=dfdrop_new).fit()
result.summary()
```

```python
result = sm.ols(formula='successful ~ usd_goal_real + usd_pledged_real + backers - 1', data=dfdrop_new).fit()
result.summary()
```

```python
result = sm.ols(formula='successful ~ backers - 1', data=dfdrop_new).fit()
result.summary()
```

```python
result = sm.ols(formula='usd_pledged_real ~ usd_goal_real + backers', data=dfdrop_new).fit()
result.summary()
```

```python
result = sm.ols(formula='usd_pledged_real ~ backers - 1', data=dfdrop_new).fit()
result.summary()
```

```python
## The above regression analysis on the full dataset (before dropping outliers) shows slight correlation
## between success and the number of backers (Adj. R-sq of only 0.038). It shows a much stronger
## correlation between amount pledged and the number of backers (Adj. R-sq of 0.601). It shows a 
## weak negative correlation between success and the magnitude of the goal (Adj. R-sq of only 0.022).
```

```python
## Drop goal outliers on high side (above 1.96 standard deviations above the mean, which turns out 
## to be above $2,000,000).
dfdrop_outliers = dfdrop_new
meangoal = dfdrop_outliers['usd_goal_real'].mean()
stdgoal = dfdrop_outliers['usd_goal_real'].std()
toprange = meangoal + stdgoal*1.96

dfdrop_outliers = dfdrop_outliers.drop(dfdrop_outliers[dfdrop_outliers['usd_goal_real']>toprange].index)

dfdrop_outliers.head()
```

```python
dfdrop_outliers.describe()
```

```python
dfdrop_outliers.corr()
```

```python
result = sm.ols(formula='successful ~ usd_goal_real + usd_pledged_real + backers + percent -1', data=dfdrop_outliers).fit()
result.summary()
## The above formula gave the highest Adj. R-sq of several formulations I tried, but of only 0.041.
## This is higher than that obtained before eliminating goal outliers.
```

```python
result = sm.ols(formula='usd_pledged_real ~ backers - 1', data=dfdrop_outliers).fit()
result.summary()
## This regression shows the same correlation between amount pledged and number of backers
## as before dropping goal outliers.
```

```python
## Sort goals into bins.
bins = [0,2,100,500,1000,1500,2000,3000,4000,5000,7500,10000,15000,30000,45000,60000,100000,500000,1000000,2000000]
group_names = ['0-2','2-100','100-500','500-1K','1K-1.5K','1.5K-2K','2K-3K','3K-4K','4K-5K','5K-7.5K','7.5K-10K','10K-15K','15K-30K','30K-45K','45K-60K','60K-100K','100K-500K','500K-1M','1M-2M']
dfdrop_outliers['goalbins'] = pd.cut(dfdrop_outliers['usd_goal_real'], bins, labels=group_names)
dfdrop_outliers.head()
```
